```python
import pandas as pd
import urllib.parse

limit = urllib.parse.quote("date_extract_y(grade_date) = 2018")
food_scores = pd.read_csv("https://data.cityofnewyork.us/resource/43nn-pn8j.csv?$where=" + limit)

food_scores.head()
```

```python
pd.set_option('display.max.columns', None)
pd.set_option('display.max.rows',None)
pd.set_option('display.max_colwidth', -1)
food_scores.head()
```

```python
food_scores['cuisine_description'].value_counts()
```

```python
#What is the median score for the 'Bakery' cuisine type?
bakery = food_scores[['cuisine_description','score']].groupby('cuisine_description').median()
bakery
```

```python
#Q3: Which cuisine type has the best (i.e., the lowest) median score? (Hint: use .sort_values)
bakery.sort_values('score')
```

```python
#Q4: Which borough (here, 'boro') has the highest (worse) mean score?
boro = food_scores[['boro','score']].groupby('boro').mean()
boro.sort_values('score')
```

```python
#Q5: Which borough has the highest count of 'A' grades?
boro = food_scores[['boro','grade']].groupby('boro')
boro['grade'].value_counts()
```

```python
#Q6: What's the grade most '02B' violations receive?
grade = food_scores[['grade','violation_code']].groupby('violation_code')
grade['grade'].value_counts()
```

```python
#Q7 What's the median score for '04L' violations?
o4l = food_scores[['score','violation_code']].groupby('violation_code')
o4l.median()
```

```python
#Q9 What is the median score for restaurants where the 'critical_flag' value was 'Critical'?
crit = food_scores[['score','critical_flag']].groupby('critical_flag')
crit.median()
```

```python
# Q10: What is the median score for restaurants receiving a 'C' grade?
cgrade = food_scores[['grade','score']].groupby('grade')
cgrade.median()
```
