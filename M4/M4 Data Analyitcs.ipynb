## Creating a simulation of per capita daily water use
#### Now that we know the mean [159.425] and standard deviation [31.58709190476388] of the average per-capita daily water use over this time period, we can create a simulation of per capita daily water use over this time frame. 
#### I will generate a unique, random value for each day in my 40 year long data set ie: 14610 values (as shown below)

```python
40*365.25
```

### Using numpy's random package, I will create an array that 
#### * contains 14610 values, 
#### * is normally distributed, 
#### * with the mean [159.425] and standard deviation [31.58709190476388] I figured out for per capita daily water use. 

```python
simulated_daily_use_array = np.random.normal(loc=159.425,scale=31.58709190476388,size=14610)
```

##### Let's double check that simulated_daily_use_array is actually a numpy array

```python
type(simulated_daily_use_array)
```

##### Now, I'll create a list that has the exact same values, but isn't an array, but rather a list

```python
simulated_daily_use_list = simulated_daily_use_array.tolist()
```

```python
simulated_daily_use_list
```

```python
type(simulated_daily_use_list)
```

## How efficient are different methods?
### I will use three different methods provided by the Professor to find the square root of each simulated water use, for the water use over 39 years for 1000 New Yorkers!
### We will use timeit to figure out what is quickest.

# Code for Profiling

```python
from timeit import timeit 
import math

def sqrt_1():
    # This one uses a loop.  Slow!
    sqrt_1 = []
    for i in range(0, len(simulated_daily_use_list)-1): 
        sqrt_1.append(math.sqrt(simulated_daily_use_list[i]))
    return(sqrt_1)
def sqrt_2():
    # This one uses list comprehension. Better!
    return([math.sqrt(x) for x in simulated_daily_use_list])

def sqrt_3():
    # This one uses numpy.  Best!
    return(np.sqrt(simulated_daily_use_array))

print("Time for loop: " + str(timeit(sqrt_1, number = 1000)))
print("Time for list comprehension: " + str(timeit(sqrt_2, number = 1000)))
print("Time for numpy vectorized operation: " + str(timeit(sqrt_3, number = 1000)))
```

### The results I got were:
#### *Time for loop: 2.835411075997399
#### *Time for list comprehension: 1.4978404169960413
#### *Time for numpy vectorized operation: 0.034351045003859326
### The point of this demostration is to show the power and value of analyzing data in numpy, which is more efficient.
