### This data is a monthly U.S. reported precipitation amounts in hundredths of inches (ex 100 is 1.00 inches) generated from the GTS metar(hourly) and synoptic(6-hourly)observations for selected cities based on the Weekly Weather and Crop Bulletin station list. You can find the data  <a href="https://catalog.data.gov/dataset/climate-prediction-center-cpc-monthly-u-s-selected-cities-precipitation-summary"> here.</a>

## Other data you may find helpful while running through this data
TEMPERATURES REPORTED IN WHOLE DEGREES FAHRENHEIT
* -99 REPRESENTS A MISSING TEMPERATURE VALUE

PRECIPITATION REPORTED IN HUNDREDTHS OF INCHES
* -999 REPRESENTS A MISSING PRECIPITATION VALUE

* ATMP:	WEEKLY OR MONTHLY AVERAGE TEMPERATURE
* BKSTN:	BLOCK STATION
* EPCP:	TOTAL PRECIPITATION INCLUDING ESTIMATES FOR THOSE PERIODS OF TIME WHEN NO AMOUNTS WERE REPORTED.
* MAXT:	DAILY MAXIMUM TEMPERATURE
* MINT:	DAILY MINIMUM TEMPERATURE
* NPCP:	WEEKLY OR MONTHLY NORMAL PRECIPITATION
* NTMP:	WEEKLY OR MONTHLY NORMAL TEMPERATURE
* PFLAG	CODE: INDICATING THE ENDING TIME THAT PRECIPITATION IS SUMMARIZED
    * 1-0300Z 2-0600Z 3-0900Z 4-1200Z 5-1500Z
    * 6-1800Z 7-2100Z 8-2400Z 9-0600Z ON SUBSEQUENT DATE

* RPCP:	TOTAL REPORTED PRECIPITATION

* WCHARS:	MOST SIGNIFIGANT WEATHER REPORTED DURING EACH 3 HOUR PERIOD
  * A - HAIL
  * B - BLOWING SNOW
  * C - SNOW CRYSTALS
  * D - DUST OR SAND
  * E - RIME FOG
  * F - FOG
  * G - GROUND FOG
  * H - HAZE
  * I - DIAMOND DUST
  * J - DRIFTING SNOW
  * K - SMOKE
  * L - DRIZZLE
  * M - MIST
  * N - NOT USED
  * O - NO SIGNIFIGANT
  * P - ICE PELLETS
  * Q - SQUALLS
  * R - RAIN
  * S - SNOW
  * T - THUNDERSTORMS
  * U - LIGHTNING
  * V - VIRGA
  * W - SHOWERS
  * X - DISTANT PRECIPITATION
  * Y - FUNNEL CLOUD(S)
  * Z - FREEZING PRECIPITATION

FILE (dly_global.txt)	DAILY GLOBAL SUMMARY FILE
* FORMAT: FIRST LINE IS THE DATE OF DATA
* DATA:
  * MAXT - 3 CHARACTERS
  * MINT - 3 CHARACTERS
  * EPCP - 4 CHARACTERS
  * RPCP - 4 CHARACTERS
  * PFLAG - 1 CHARACTER
  * WCHAR - 8 CHARACTERS
  * BKSTN - 5 CHARACTERS

* FILES: (wkly_global.txt and mnly_global.txt)	WEEKLY & MONTHLY GLOBAL SUMMARY FILE, RESPECTIVELY.

* FORMAT: FIRST LINE IS THE DATE OF DATA

* Data:
  * BKSTN - 5 CHARACTERS
  * ATMP - 3 CHARACTERS
  * NTMP - 3 CHARACTERS
  * RPCP - 4 CHARACTERS
  * EPCP - 4 CHARACTERS
  * NPCP - 4 CHARACTERS

## The first thing you notice when you access the data, is that it isn't in a normal, `csv` file type. instead, it is using fixed widths (as indicated int he field above) to indicate what data belongs in what row and column. We will try to read this data into our python file below, and print out a preview, after removing the first few lines which contain.... less rectangular data.

```python
import pandas as pd
precip_df = pd.read_fwf('/Users/leidner/Downloads/Percipitation.txt',
                        skiprows=10,
                        header=None)
precip_df.head(20)
```

```python
precip_df.tail(20)
```

### Not bad, though some cleanup is required. First we'll drop the last row, which never should have been loaded in. We see from the cell above the the row is `249` so...

```python
precip_df = precip_df.drop(249)
precip_df.tail(5)
```

### Better. But we still have some columns that did not import properly. Columns 1 and 10 are both doubling up on data. Let's see what we can do about that.

```python
precip_df['city'] = precip_df[1].str[:-1]
precip_df['pflag'] = precip_df[1].str[-1]
```

```python
precip_df.head(5)
```

### We've succesfully split column 1 into two columns with the correct columns names. Now lets drop the redundant column 1

```python
precip_df = precip_df.drop(columns=[1])
precip_df.head()
```

### Now we'll do the same for column 10

```python
precip_df['year_to_date_amt'] = precip_df[10].str[:4]
precip_df['year_to_date_dep_nml'] = precip_df[10].str[4:]

precip_df = precip_df.drop(columns=[10])
precip_df.head()
```

### OKAY. Looks clean enough to me. Now, all we have to do are supply column names and maybe re-order things so that the city and pflag come as the second and third columns, respectively, like they did in the original file.

```python
precip_df.rename(columns={0:'state',
                          2: "call_ltrs",
                          3: "month_tot_amt",
                          4: "month_dep_nml",
                          5: "24_hr_max",
                          6: "days_gt_1",
                          7: "days_gt_50",
                          8: "season_to_date",
                          9: "season_dep_nml",
                          11: "year_to_date_pct_nml"},inplace=True)
precip_df.head()
```

### Renamed, but not reordered

```python
precip_df = precip_df[["state",
                       "city",
                       "pflag",
                       "call_ltrs",
                       "month_tot_amt",
                       "month_dep_nml",
                       "24_hr_max",
                       "days_gt_1",
                       "days_gt_50",
                       "season_to_date",
                       "season_dep_nml",
                       "year_to_date_amt",
                       "year_to_date_dep_nml",
                       "year_to_date_pct_nml"]]
precip_df.head()
```

### Now that our data set is importaed and cleaned up, we can start analyzing it! 
### We'll start by identifying the city with the highest and lowest year_to_date_pct_nml. We'll `sort` the data by that column first, then we can just look at the first and the last row. to see the highest number.

```python
precip_df = precip_df.sort_values('month_dep_nml')
precip_df.head()
```

```python
precip_df.iloc[0]
```

### Ft Lauderdale, Florida has 8 times less of its percepitation this year to date? Florida must be *much* drier than normal this year. To find the last row, and thus the hieghtest number, we can remember what our list length was from above or use the `len` function

```python
len(precip_df)
```

```python
precip_df.iloc[248]
```

### BEAUMONT, TX is very, * very* wet this year, with 30 times its normal precipitation. Was there flooding?

### We've done a lot of work to re-format our data. Let's write this neatned up dataset to a more readable file.

```python
precip_df.to_csv('/Users/leidner/Downloads/precipitation_data.csv')
```
